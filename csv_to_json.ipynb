```python
import pandas as pd
data = pd.read_csv('film_cleaned.csv')

import csv  
import json  
import math

dict = {
    "Year" : [elt for elt in data["Year"]],
    "Length" : [elt for elt in data["Length"]],
    "Title" : [elt for elt in data["Title"]],
    "Subject" : [elt for elt in data["Subject"]],
    "Actor" : [elt for elt in data["Actor"]],
    "Actress" : [elt for elt in data["Actress"]],
    "Director" : [elt for elt in data["Director"]],
    "Popularity" : [elt for elt in data["Popularity"]],
    "Awards" : [elt for elt in data["Awards"]]
}

out = json.dumps(dict, separators=(',',':'))
f = open( 'film_cleaned.json', 'w')  
f.write(out)
f.close()

dict
```

```python
import pandas as pd
data = pd.read_csv('film_cleaned.csv')

import csv  
import json  
import math

# Open the CSV  
f = open( 'film_cleaned.csv', 'rU' )  
# Change each fieldname to the appropriate field name. I know, so difficult.  
reader = csv.DictReader( f, fieldnames = ( "Year", "Length", "Title", "Subject", "Actor", "Actress", "Director", "Popularity", "Awards", "*Image" ))  
# Parse the CSV into JSON  
out = json.dumps( [ row for row in reader ] )  
print("JSON parsed!") 
# Save the JSON  
f = open( 'summary.json', 'w')  
f.write(out)
f.close()
print("JSON saved!")
```

```python
import pandas as pd
data = pd.read_csv('film_cleaned.csv')

import csv  
import json  
import math

dict = {}

for i in range(len(data)):
    dict[data["Title"].iloc[i]] = {"Year" : int(data["Year"].iloc[i]),
                                   "Length" : int(data["Length"].iloc[i]), 
                                   "Subject" : data["Subject"].iloc[i], 
                                   "Actor" : data["Actor"].iloc[i],
                                   "Actress" : data["Actress"].iloc[i],
                                   "Director" : data["Director"].iloc[i],
                                   "Popularity" : int(data["Popularity"].iloc[i]),
                                   "Awards" : data["Awards"].iloc[i]}
    
out = json.dumps(dict, separators=(',',':'))
f = open( 'summary.json', 'w')  
f.write(out)
f.close()

dict
```
